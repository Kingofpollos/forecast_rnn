```python
import yfinance as yf
import pandas as pd
import numpy as np

import  libs as l
```

```python
activos = [
    # Acciones de grandes empresas tecnológicas (EE.UU.)
    "AAPL",  # Apple Inc.
    "MSFT",  # Microsoft Corporation
    "GOOGL",  # Alphabet Inc. (Google)
    "AMZN",  # Amazon.com Inc.
    "META",  # Meta Platforms Inc. (Facebook)

    # Índices bursátiles
    "^GSPC",  # S&P 500
    "^DJI",  # Dow Jones Industrial Average
    "^IXIC",  # NASDAQ Composite
    "^VIX",  # Índice de volatilidad (Fear Index)

    # Empresas globales fuera de EE.UU.
    "TSLA",  # Tesla, Inc.
    "BABA",  # Alibaba Group Holding Ltd. (China)
    "NIO",  # NIO Inc. (Vehículos eléctricos de China)
    "VALE",  # Vale S.A. (Minería en Brasil)

    # ETFs y Fondos Cotizados
    "SPY",  # SPDR S&P 500 ETF Trust (EE.UU.)
    "QQQ",  # Invesco QQQ Trust (Seguimiento NASDAQ-100)
    "VWO",  # Vanguard FTSE Emerging Markets ETF
    "EWW",  # iShares MSCI Mexico ETF
    "ARKK"  # ARK Innovation ETF
]   
```

```python
def get_data(ticker, start, end):
    data = yf.download(ticker, start=start, end=end)[["Adj Close"]]
    data = [data[ticker] for ticker in data.columns]
    index, xs, ys = [], [], []
    for i in range(len(data)):
        dates = list(map(lambda x: str(x.strftime('%Y-%m-%d')), data[i].index.to_list()))
        x,y = l.split_sequence(data[i], 28, 5)
        ind = list(map(lambda x: x[0]+"-"+x[1]+"-"+ticker[i], l.split_dates(dates, 28,5)))
        x = x.reshape((x.shape[0], x.shape[1]))
        x_norm = np.apply_along_axis(l.vector_normalizer , 1, x)
        x_minmax = np.apply_along_axis(l.vector_min_max , 1, x)
        y = y.reshape((y.shape[0], y.shape[1]))
        y_norm = l.apply_normalizer(y, x_minmax)
        index.append(ind)
        xs.append(x_norm)
        ys.append(y_norm)
    return [index, xs, ys]

index, xs ,ys = get_data(activos, "2021-01-01", "2021-12-31")
```
